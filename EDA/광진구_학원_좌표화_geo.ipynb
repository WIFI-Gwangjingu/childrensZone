```python
import numpy as np
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/공모전/광진구/서울시 광진구 학원 교습소정보.csv', encoding='euc-kr')
df
```

```python
df.info()
```

```python
#필요한 column만 추출
df =df.loc[:,['학원/교습소','학원명','도로명주소','교습과정명']]
df
```

```python
#교습과정명 추출
df['교습과정명'].unique()
```

```python
#성인 대상 학원 row 삭제
df = df[~df['교습과정명'].str.contains("성인", na=False, case=False)]
df['교습과정명'].unique()
```

```python
df.info()
```

```python
#데이터프레임 주소값 추출
address = df['도로명주소']
address
```

```python
####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]

#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
#####Dataframe만들기######
address_df = pd.DataFrame({'학원/교습소':df['학원/교습소'], '교습과정명':df['교습과정명'],'학원명': df['학원명'],'인원':df['일시수용능력인원합계'],주소':df['도로명주소'],'위도':latitude,'경도':longitude})
address_df
```
