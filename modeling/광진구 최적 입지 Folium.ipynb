# 지도에 표시

```python
g_map = g.Map(location=[37.5467832,127.077811],tiles='openstreetmap', zoom_start=14ㄴ)

for i in data.index:
    lat = data.loc[i,'위도']
    lng = data.loc[i,'경도']
    name = data.loc[i,'학원명']

    marker = g.Marker([lat,lng],popup=name)
    marker.add_to(g_map)

for i in data.index:
    lat = data.loc[i,'위도']
    lng = data.loc[i,'경도']
    name = data.loc[i,'학원명']

    marker = g.Circle([lat,lng],radius=500)
    marker.add_to(g_map)

g_map
```
