```python
nw = pd.read_csv("노원구데이터프레임.csv")
nw
```

```python
data = pd.concat([gn, nw], ignore_index = True)
data
```

```python
feature = ['인원','근처학원수','근처수강생총합','근처휴게음식점수','근처횡단보도수','사고건수']
target = "어린이보호구역여부"
```

```python
X_train = data[feature]
y_train = data[target]
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

# 스케일러 생성
scaler = StandardScaler()

# 스케일 조정
X_train_sclaed = scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

# 모델 생성 및 학습 시키기
logistic = LogisticRegression(max_iter = 5000)
logistic.fit(X_train_sclaed, y_train)

# 결과 확인
# print("validation 데이터셋 정확도")
# logistic.score(X_train, y_val)
```

```python
logistic.coef_
```

```python
print("validation 데이터셋 정확도")
logistic.score(X_train_sclaed, y_train)
```

```python
model_coef = pd.Series((logistic.coef_[0]), feature).sort_values(ascending=False)
print(model_coef)

# 회귀 계수의 절대값이 클 수 록  라벨 결정에 큰 영향을 미침.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
label = list(model_coef.index)
```

```python
value = list(model_coef)
```

```python
weight = pd.DataFrame(value,label)
weight
```

```python
weight.rename(columns = {0 : 'weight'}, inplace = True)
```

```python
weight
```

```python
weight = weight.sort_values(by=['weight'], ascending=True) 
ax = weight.plot.barh(y='weight', color= '#FFCC00')
```
